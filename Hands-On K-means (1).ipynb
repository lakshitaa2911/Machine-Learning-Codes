```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
```

```python
X = np.array([[5,3], [10,15],[15,12],[24,10],[30,45],[85,70],[71,80],[60,78],[55,52],[80,91]])
type(X)
```

```python
plt.scatter(X[:,0],X[:,1], label='True Position')
```

```python
# Using the elbow method to find the optimal number of clusters mean chose the value of k
from sklearn.cluster import KMeans 
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)	#Sum of squared distances of samples to their closest cluster center.
plt.plot(range(1, 11), wcss) 
plt.title('The Elbow Method')
plt.xlabel('Number of clusters') 
plt.ylabel('WCSS')
plt.show()
#lowerst bend point is the correct value of k
```

```python
kmeans = KMeans(n_clusters=3) 
kmeans.fit(X)
```

```python
print(kmeans.cluster_centers_)
```

```python
print(kmeans.labels_)
```

```python
plt.scatter(X[:,0],X[:,1], c=kmeans.labels_)
```

```python
plt.scatter(X[:,0], X[:,1], c=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:,0] ,kmeans.cluster_centers_[:,1], color='red')
```
