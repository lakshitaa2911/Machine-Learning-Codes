### QUESTION 4

Write a python code to Implement the SVM classifier on Breast Cancer Dataset using scikit-learn. Also check the accuracy of the model. 

#### SVM Classifier on Breast Cancer Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('breast.csv')
df.head(20)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.drop(axis =1,columns ='Unnamed: 32',inplace = True)
```

```python
X = df.iloc[:, 2:].values
y = df.iloc[:, 1].values
```

```python
X
```

```python
y
```

```python
df.isnull().sum()
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.svm import SVC
classifier = SVC(kernel = 'rbf')
classifier.fit(X_train, y_train)
```

```python
# Predicting the Test set results
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
import seaborn as sn
plt.figure(figsize = (5,5))
#sn.heatmap(cm, annot=True)
sn.heatmap(cm, fmt=".3f", linewidths=.5, square = True, annot=True)
plt.xlabel('Actual label')
plt.xlabel('Actual label')
plt.ylabel('Predicted label')
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test, y_pred)*100
acc
```

#### Accuracy of the trained model is 98.24561403508771 

### QUESTION 5

Write a python code to Implement the KNN classifier on Glass Type Classification dataset using scikit-learn. Also check the accuracy of the model

##### KNN model on glass Dataset 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('glass.csv')
dataset
```

```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
sc_x.fit(dataset.drop('Type', axis=1))
scaled_features = sc_x.transform(dataset.drop('Type',axis=1))

# Convert data into data frame
dataset_updated = pd.DataFrame(scaled_features, columns = dataset.columns[:-1])
print(dataset_updated.head())
```

```python
# Now train test split one..
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(scaled_features, dataset['Type'],test_size=0.2,random_state=42)

from sklearn.neighbors import KNeighborsClassifier
Obj = KNeighborsClassifier(n_neighbors = 11)
Obj.fit(xtrain,ytrain)
pred = Obj.predict(xtest)

print(ytest,pred)
```

```python
acc_rate=[]
for i in range(1,40):
    Obj = KNeighborsClassifier(n_neighbors = i)
    Obj.fit(xtrain,ytrain)
    pred_i = Obj.predict(xtest)
    acc_rate.append(1-np.mean(pred_i !=ytest))
```

```python
error_rate=[]
for i in range(1,40):
    Obj = KNeighborsClassifier(n_neighbors = i)
    Obj.fit(xtrain,ytrain)
    pred_i = Obj.predict(xtest)
    error_rate.append(np.mean(pred_i !=ytest))
```

```python
plt.figure(figsize=(10,8))
# plt.plot(range(1,40), error_rate, color='red',linestyle='dashed',marker='0',markerfo)
plt.plot(range(1,40), acc_rate,color='purple',linestyle='dashed',marker='^')
plt.title('Accuracy vs K Value')
plt.xlabel('K')
plt.ylabel('Accuracy Rate')
```

```python
Obj = KNeighborsClassifier(n_neighbors = 1)
Obj.fit(xtrain,ytrain)
pred = Obj.predict(xtest)
print(pred)
# acc_rate.append(1-np.mean(pred_i !=ytest))
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(pred,ytest)
acc*100
```
